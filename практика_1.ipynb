```python
# 1. Распределение финальных оценок по планам на высшее образование
print("\n1. Распределение итоговых оценок по планам на высшее образование:")
sns.histplot(data=df, x="G3", hue="higher", kde=True, element="step")
plt.title("Распределение итоговых оценок по планам на высшее образование")
plt.xlabel("Финальная оценка (G3)")
plt.ylabel("Количество студентов")
plt.show()

print("Является ли распределение нормальным или отлично от него?")
print("Ответ: Распределение имеет два выраженных пика около 10 и 15 баллов, что указывает на бимодальность и значительное отклонение от нормального распределения")

print("Концентрация большинства значений")
print("Ответ: большинство значений сконцентрировано в диапазоне от 8 до 16 баллов")

print("Анализ категориальной переменной:")
print("Ответ: Распределение оценок G3 в зависимости от планов на высшее образование (higher) показывает, что студенты, планирующие получить высшее образование, имеют значительно более высокие оценки с пиком около 15 баллов, тогда как у студентов без планов на высшее образование пик распределения находится около 10 баллов")

# 2. Распределение времени на учебу по полу
print("\n2. Время на учебу в зависимости от пола:")
sns.histplot(data=df, x="studytime", hue="sex", multiple="dodge")
plt.title("Время на учебу в зависимости от пола")
plt.xlabel("Время на учебу (1-мало → 4-много)")
plt.ylabel("Количество студентов")
plt.show()

print("Является ли распределение нормальным или отлично от него?")
print("Ответ: Распределение дискретное с четырьмя категориями, что полностью отличается от нормального непрерывного распределения")

print("Концентрация большинства значений")
print("Ответ: большинство значений сконцентрировано на уровне 2 часа в неделю")

print("Анализ категориальной переменной:")
print("Ответ: Распределение времени на учебу (studytime) в зависимости от пола (sex) показывает, что девушки чаще выбирают средние значения (2-3 часа), тогда как юноши более равномерно распределены по всем категориям, включая крайние значения 1 и 4 часа")

# 3. Распределение потребления алкоголя по выходным по полу
print("\n3. Потребление алкоголя в выходные по полу:")
sns.histplot(data=df, x="Walc", hue="sex", multiple="stack", discrete=True)
plt.title("Потребление алкоголя в выходные по полу")
plt.xlabel("Уровень алкоголя (1-низкий → 5-высокий)")
plt.ylabel("Количество студентов")
plt.show()

print("Является ли распределение нормальным или отлично от него?")
print("Ответ: Распределение дискретное с ярко выраженной правосторонней асимметрией, что значительно отличается от нормального распределения")

print("Концентрация большинства значений")
print("Ответ: большинство значений сконцентрировано на низких уровнях потребления алкоголя (1-2 балла)")

print("Анализ категориальной переменной:")
print("Ответ: Распределение потребления алкоголя по выходным (Walc) в зависимости от пола (sex) показывает, что юноши значительно чаще употребляют алкоголь на высоких уровнях (4-5 баллов), тогда как девушки преобладают в категориях низкого потребления (1-2 балла)")
```

```python
corr_matrix = df.select_dtypes(include=['int64', 'float64']).corr()

# Визуализация
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Корреляционная матрица (числовые переменные)")
plt.show()
```

Какие переменные не коррелируют?

> возраст, здоровье, семейные отношения

Какие переменные имеют сильную положительную корреляцию?

> оценки G1, G2, G3 (0.80-0.90)

Какие переменные имеют сильную слабую корреляцию?

> age и studytime: (-0.00). famrel и оценки: 0.02-0.05, семейные отношения почти не влияют на оценки
